# Ejemplo generación de texto usando modelos pre entrenados (GPT-2)

Usando el modelo GPT-2 de libre acceso proveniente de [Hugging Face](https://huggingface.co/docs/transformers/model_doc/gpt2) generaremos oraciones a partir de una secuencia de palabras dada.

El primer paso es instalar la librería necesaria, en python ejecutamos el siguiente código.

```python
pip install transformers
```

## Carga del modelo

```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2')
```

## Generación de texto a partir de una secuencia dada

El modelo GPT-2 tiene mejor rendimiento cuando el idioma utilizado es inglés. Observemos algunos ejemplos

```python
set_seed(123) # Ajustamos la semilla para que puedan replicarse los resultados en distintos equipos
generator('A transformer is', # Secuencia de palabras inicial
          max_length=20,      # Número máximo de palabras
          num_return_sequences=3)  # Número máximo de oraciones generadas
```

```python
generator('Un transformador es',
          max_length=20,
          num_return_sequences=3)
```
